```python
import math
def euclidean_distance(t1,t2):
  # takes in two tuples
  return math.sqrt((t1[0]-t2[0])**2 + (t1[1]-t2[1])**2 + (t1[2]-t2[2])**2)
```

```python
three_2_one = {}
```

```python
with open("6wps.pdb") as f:
    lines = f.readlines()
```

```python
with open("123.txt") as f:
    symbols = f.readlines()
```

```python
for sym in symbols:
    sym = sym.split()
    one = sym[0]
    three = sym[1].upper()
    three_2_one[three] = one
```

```python
three_2_one
```

```python
# we need alpha carbon
# ATOM      2  CA  ALA A  27     193.200 155.103 203.437  1.00 47.60           C   
# So we want to extract lines like the one above
coord = {}
for line in lines:
    words = line.strip("\n").split()
#     atom = words[0]
#     alpha = words[2]
    if "ATOM" in words and "CA" in words:
        print(words)
        x = float(words[6])
        y = float(words[7])
        z = float(words[8])
        chain = words[4]
        symbol = three_2_one[words[3]]
        seq = words[5]
        name = symbol + chain + seq+"-6wps"
        coord[name] = (x,y,z)
```

```python
def get_coordinates():
    coord = {}
    for line in lines:
        words = line.strip("\n").split()
    #     atom = words[0]
    #     alpha = words[2]
        if "ATOM" in words and "CA" in words:
            print(words)
            x = float(words[6])
            y = float(words[7])
            z = float(words[8])
            chain = words[4]
            symbol = three_2_one[words[3]]
            seq = words[5]
            name = symbol + chain + seq+"-6wps"
            coord[name] = (x,y,z)
    return coord
```

```python
distance_dict = {}
```

```python
alphas = list(coord.keys())
euclidean_distance(coord[alphas[0]],coord[alphas[1]])
# alphas
```

```python
for i in range(len(alphas)):
    # only consider the atoms ahead of this
    atoms = alphas[i+1:]
    for atom in atoms:
        pair = alphas[i]+"<--->"+atom
        distance_dict[pair] = euclidean_distance(coord[alphas[i]],coord[atom])
```
